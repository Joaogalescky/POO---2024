# Exercício 02

<p style="text-align:justify;">
Crie uma classe que efetue a leitura de um valor inteiro e outro real, exibindo-os no console.
</p>


```java
import java.util.Scanner;

public class teste {
    public static void main(String[] args) {
        Scanner entrada = new Scanner(System.in);

        System.out.print("Digite um valor inteiro: ");
        int valorInteiro = entrada.nextInt();

        System.out.print("Digite um valor real: ");
        double valorReal = entrada.nextDouble();

        System.out.printf("Valor inteiro digitado: %d\n", valorInteiro);
        System.out.printf("Valor real digitado: %f\n", valorReal);

        entrada.close();
    }
}
```
